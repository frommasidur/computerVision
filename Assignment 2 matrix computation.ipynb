Two matrices are given one is a bigger (W) another square of size 3X3 (K). Now we have to
implement the following operations to get the resultant matrix R.

```python
import numpy as np
```

```python
wSize = int(input("Enter the size of w matrix: "))
kSize = int(input("Enter the size of k matrix: "))
rSize = wSize-kSize
```

```python
W = np.zeros((wSize,wSize))
K = np.zeros((kSize,kSize))
R = np.zeros((rSize,rSize))
```

```python
for i in range(wSize):
  for j in range(wSize):
    W[i][j] = (i+1)*(j*2)
for i in range(kSize):
  for j in range(kSize):
    K[i][j] = i*j
```

```python
print("Matrix W")
for i in range(wSize):
  for j in range(wSize):
    print(W[i][j],end="  ")
  print()
print("Matrix K")
for i in range(kSize):
  for j in range(kSize):
    print(K[i][j],end=" ")
  print()
```

```python
sum = 0
for i in range(rSize):
  for j in range(rSize):
    for k in range(0,kSize):
      for l in range(0,kSize):
        sum += W[k+i][l+j]*K[k][l]
    R[i][j] = sum
    sum = 0
print("R matrix: ")
for k in range(0,rSize):
      for l in range(0,rSize):
        print(R[k][l],end=" ")
      print()
```
